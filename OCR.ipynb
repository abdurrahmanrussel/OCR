```python
# ‚úÖ Install dependencies
!pip install pytesseract pdf2image PyPDF2
!apt-get install -y poppler-utils tesseract-ocr tesseract-ocr-ben

# ‚úÖ Upload scanned Bangla PDF
from google.colab import files
uploaded = files.upload()
pdf_file = list(uploaded.keys())[0]

# ‚úÖ Import libraries
import pytesseract
from pdf2image import convert_from_path
from PyPDF2 import PdfReader  # ‚úÖ Use PdfReader instead of PdfFileReader

# ‚úÖ Count total pages in the PDF
with open(pdf_file, "rb") as f:
    reader = PdfReader(f)
    total_pages = len(reader.pages)  # ‚úÖ Updated way to get total pages

print(f"üìÑ Total Pages Detected: {total_pages}")

# ‚úÖ Process all pages one by one (memory-efficient)
text = ""
for page_num in range(1, total_pages + 1):
    try:
        images = convert_from_path(pdf_file, dpi=150, first_page=page_num, last_page=page_num)
        img = images[0]
        page_text = pytesseract.image_to_string(img, lang='ben')  # Bengali OCR
        text += page_text + "\n"
        print(f"‚úÖ OCR completed for page {page_num}/{total_pages}")
    except Exception as e:
        print(f"‚ö†Ô∏è Error on page {page_num}: {e}")

# ‚úÖ Save full extracted text
with open("bangla_book_full_ocr.txt", "w", encoding="utf-8") as f:
    f.write(text)

# ‚úÖ Provide a download link
files.download("bangla_book_full_ocr.txt")
```

```python
##OCR English
# ‚úÖ Install required packages
!pip install pytesseract pdf2image
!apt-get install -y poppler-utils tesseract-ocr

# ‚úÖ Upload your scanned English PDF
from google.colab import files
uploaded = files.upload()
pdf_file = list(uploaded.keys())[0]

# ‚úÖ OCR: Convert PDF pages to images and extract English text
from pdf2image import convert_from_path
import pytesseract

text = ""
images = convert_from_path(pdf_file, dpi=300)  # High DPI for better accuracy

for i, img in enumerate(images):
    page_text = pytesseract.image_to_string(img, lang='eng')  # English OCR
    text += page_text + "\n"
    print(f"‚úÖ OCR done for page {i+1}/{len(images)}")

# ‚úÖ Save the full extracted text
with open("english_ocr_output.txt", "w", encoding="utf-8") as f:
    f.write(text)

# ‚úÖ Download the extracted text
files.download("english_ocr_output.txt")
```
